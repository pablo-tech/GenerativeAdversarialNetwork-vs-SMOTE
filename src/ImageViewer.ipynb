```python
project_name = 'keras-gan-accuracy'
path_to_project = "/home/ec2-user/SageMaker/Rosenblatt-AI/CS230/src/images/"
full_path = path_to_project + project_name
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
import glob
import os
from os import listdir
from os.path import isfile, join

file_names = [f for f in listdir(full_path) if isfile(join(full_path, f))]
print(file_names)

qualified_files = glob.glob(full_path + '/' + '*') # * means all if need specific format then *.csv
sorted_files = sorted(qualified_files, key=os.path.getctime, reverse=True)
```

```python
target_count = 8
image_count = len(sorted_files)
one_every = int(image_count/target_count)
if one_every<2:
    one_every = 2

sample_images = []
best_image = sorted_files[0]
sample_images.append(best_image) 

total_sampled = 2
for i in range(1, image_count): 
    file_path = sorted_files[i]
    this_time = (i%int(one_every-1) == 0)
    if this_time and total_sampled<target_count:
        sample_images.append(file_path)
        total_sampled += 1
    i += 1

first_image = sorted_files[image_count-1]
sample_images.append(first_image) 
```

```python
plt.figure(figsize=(12,12))

i=0
for image in sample_images:
    plt.subplot(4, 4, i+1)
    img = mpimg.imread(image)
    plt.imshow(img, cmap='gray')
    plt.axis('off')
    i += 1

plt.tight_layout()
plt.show()
print("best_image=", best_image)
```
